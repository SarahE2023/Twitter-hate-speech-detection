We have an imbalanced dataset here. Later we will deal with these imbalances by using classifiers that work on both unbalanced and balanced data set. However, for now let us focus on cleaning our dataset.


1.   Removing punctuations
2.   Removing Uppercasing
3.   Removing stopwards
4.   Stemming vs lemmatization






```python
PUNCT_TO_REMOVE = string.punctuation
def remove_punctuation(tweet):
  return tweet.translate(str.maketrans('', '', PUNCT_TO_REMOVE))

hate["tweets_without_punctuation"] = hate["tweet"].apply(lambda tweet: remove_punctuation(tweet))
```

Removing Uppercasing and stopwords

```python
hate["tweet_lower"] = hate["tweets_without_punctuation"].str.lower()
```

```python
#listing stopwords
from nltk.corpus import stopwords
", ".join(stopwords.words('english'))
```

```python
STOPWORDS = set(stopwords.words('english'))
def remove_stopwords(tweet_lower):
    return " ".join([word for word in str(tweet_lower).split() if word not in STOPWORDS])

hate["cleaned_tweets"] = hate["tweet_lower"].apply(lambda tweet_lower: remove_stopwords(tweet_lower))
```

```python
from collections import Counter
cnt = Counter()
for text in hate["cleaned_tweets"].values:
    for word in text.split():
        cnt[word] += 1

cnt.most_common(10)
```

There are two major ways of reducing words to their stems;
1.   Stemming
2.   Lemmitisation

We are going to be exploring both options to observe its impact on our dataset
